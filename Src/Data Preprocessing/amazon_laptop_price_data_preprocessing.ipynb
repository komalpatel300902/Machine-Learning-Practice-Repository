#  price Column

```python
def dollor_remover(value):
    return value[1::]

df["price"] = df['price'].apply(dollor_remover)
```

```python
def striping(value):
    return value.strip()

df["price"] = df["price"].apply(striping)
```

```python
def coma_remover(value):
    s = ""
    for x in value:
        if x != ",":
            s += x
    return s


df["price"] = df["price"].apply(coma_remover)
```

```python
df["price"] = df["price"].astype(float)
```

# For OS column in dataframe


```python
df["OS"].value_counts()
```

```python
def win_identifier(value):
    if type(value) == float:
        return np.nan;
    elif "Win" in value:
        return "Win"
    return value
df["OS"] = df["OS"].apply(win_identifier)
```

```python
df["OS"].value_counts()
```

```python
def mac_identifier(value):
    if type(value) == float:
        return np.nan;
    elif "mac" in value or "Mac" in value:
        return "Mac"
    return value
df["OS"] = df["OS"].apply(mac_identifier)
```
